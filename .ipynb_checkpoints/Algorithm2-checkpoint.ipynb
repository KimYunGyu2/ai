```python
a = int(input('첫번째 정수 : '))
b = int(input('두번째 정수 : '))

if a > b :
    print(f'둘중 더 큰 수는 {a} 이다.')
    
elif a == b :
    print('a와 b는 같다')
else :
    print(f'둘중 더 큰 수는 {b} 이다.')
```

```python
# 두 변수의 데이터를 교환 (파이썬만 가능)

a = 1
b = 2

print('a = ',a)
print('b = ',b)


a, b = b, a

print('a = ',a)
print('b = ',b)
```

```python
# 두 변수의 데이터를 교환 (공용)

a = 1
b = 2

print('a = ',a)
print('b = ',b)

t = a
a = b
b = t

print('a = ',a)
print('b = ',b)
```

```python
# 합계값을 계산하는 알고리즘

첨자 index  0  1  2  3  4
데이터     12 13 11 14 10

데이터들의 합계를 구하는 가장 단순한 알고리즘은 sum

arr[0]+arr[1]+arr[2]+arr[3]+arr{4}

로도 표현할수 있다. 그렇지만 배열의 개수가 늘어나면 사용하기 힘들다
먼저 반복 구조로 변형하자.

arr[0] -> sum
sum + arr[1] -> sum
sum + arr[2] -> sum
sum + arr[3] -> sum
sum + arr[4] -> sum
```

```python
arr = [12, 13, 11, 14, 10]

sum = 0

for i in arr:
    sum = sum + i
    
print(sum)
```

```python
# 최대값을 찾는 알고리즘

# arr = [12, 13, 11, 14, 10]

# 첨자 index  0  1  2  3  4
# 데이터     12 13 11 14 10

arr = [12, 13, 11, 14, 10]

max = arr[0]

for i in arr:
    if i > max:
        max = i
    
print(max)
```
