```python
from selenium import webdriver

from bs4 import BeautifulSoup as bs
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.melon.com/chart/index.htm')
```

```python
html = driver.page_source
soup = bs(html)
soup
```

```python
len(soup.select('tr.lst50'))
```

```python
songs = soup.select('tr.lst50')
```

```python
song = songs[0]
song
```

```python
title = songs[0].select('div.rank01')[0].text.strip()
title
```

```python
singer = songs[0].select('div.rank02')[0].text.strip()
singer_name = singer[0:len(singer)//2]
singer_name
```

```python
song_data = []
rank = 1

songs = soup.select('tr.lst50')  # 50곡 전체

for song in songs:
    title = song.select('div.rank01')[0].text.strip()
    singer = song.select('div.rank02')[0].text.strip()
    singer_name = singer[0:len(singer)//2]
    song_data.append([rank, title, singer_name])
    rank += 1
song_data
```

```python
import pandas as pd
df = pd.DataFrame(song_data, columns = ['순위','타이틀','가수'])
df
```

```python
df.to_excel('Melon.xlsx', index=False)
```
